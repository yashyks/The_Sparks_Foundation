City wise profit

```python
City_Profit = df.groupby('City').sum()['Profit']
print("City wise Profit in Ascending Order: \n", City_Profit)
```

```python
city_sorted1 = City_Profit.sort_values(ascending = False).head()
#plot
fig,axes = plt.subplots(1,2, figsize = (15,5))
city_sorted1.plot(kind ='bar', ax = axes[0])
axes[0].tick_params(labelrotation = 90)
axes[0].set_title("Maximum City Profit (Top 5)")
city_sorted2 = City_Profit.sort_values(ascending = True).head()
city_sorted2.head().plot(kind='bar',ax = axes[1])
plt.xticks(rotation = 90)
plt.title("Minimum City Profit (Bottom 5)")
```

State wise profit

```python
State_Profit = df.groupby('State').sum()['Profit']
print("State wise profit:\n",State_Profit)
```

```python

State_Profit.sort_values(ascending = False)
#plot
plt.figure(figsize =(15,4))
sns.barplot(x = State_Profit.index, y = State_Profit.values)
plt.xticks(rotation =90)
plt.title("State Wise Profit")
```

Region wise

```python
Region_profit = df.groupby('Region').sum()['Profit']
print("Region wise profit:\n",Region_profit)
```

```python
#Display Region wise profit in maximum to minimum order
Region_profit.sort_values(ascending = False)
#Plot
plt.pie(Region_profit, labels = Region_profit.index, autopct= "%.2f", explode = (0.05,0.05,0.05,0.05),radius=1.3)
fig= plt.gcf()
```
